```python
from glob import glob

files = sorted(glob(f'{droot}/PRI_co2gridfed_output1.vGV7.sGV7/*o.co2.ts'))

def parse_filename(name: str):
    
    # first letter
    method_def = {'H': 'insitu', 'F': 'flask'}
    assert name[0] in method_def, f'method: unknown code: {name}'
    
    # second letter
    type_def = {'b': 'surface', 'r': 'surface', 's': 'surface', 'o': 'shipboard', 
                'a': 'unknown', 'c': 'unknown', 'm': 'unknown', 't': 'unknown', 
                'u': 'unknown'}
    assert name[1] in type_def, f'type: unknown code: {name}'
    
    # capital letters, station code
    station = ''
    for n in range(2, len(name)):
        c = name[n]
        if c.isupper() or c.isnumeric() or c in ['_']:
            station += c
        else:
            break
    
    labcode_def = {'c': 'NOAA', 'a': 'CSIRO', 'k': 'SIO_CDK', 's': 'SIO_O2',
                   'l': 'LSCE', 'x': 'TU, NIWA, or SAWS', 
                   'n': 'unknown', 'm': 'unknown', 'd': 'unknown', 'e': 'unknown',
                   'b': 'unknown', 'u': 'unknown', 'j': 'unknown', 'r': 'unknown'
                  }
    
    if name[n] not in labcode_def:
        print(name[n])
        print(station)
        print(f'lab: unknown code: {name}')
        raise
    
    attrs = {
        'method': method_def[name[0]],
        'type': type_def[name[1]],
        'station': station,
        #'institution': labcode_def[name[5]],
    }
    return attrs



for f in files:
    basename = os.path.basename(f)
    attrs = parse_filename(basename)
    
```
