```python
import matplotlib.pyplot as plt
import xarray as xr
import util
```

```python
mwco2 = 44.0
```

```python
INPUTDATA = '/glade/p/cesmdata/cseg/inputdata'
co2flux_fuel_file = f'{INPUTDATA}/atm/cam/ggas/emissions-cmip6_CO2_anthro_surface_175001-201512_fv_0.9x1.25_c20181011.nc'
#aircraft_specifier             = 'ac_CO2 -> ac_CO2_filelist_175001-201512_fv_0.9x1.25_c20181011.txt'
ds = xr.open_dataset(co2flux_fuel_file).load()
ds
```

```python
ds.CO2_flux
```

```python
area = util.compute_grid_area(ds)
```

```python
with xr.set_options(keep_attrs=True):
    SFCO2_global = (ds.CO2_flux * area).sum(['lat', 'lon']) * 86400. * 365. * 1e-15 * 1e3
    SFCO2_global.attrs['units'] = 'Gt CO$_2$ yr$^{-1}$'
SFCO2_global    
```

```python
SFCO2_global.sel(time=slice('1980', '2015')).plot()
```

```python
catalog = util.curate_flux_products().open_catalog()
catalog
```

```python
ds_odiac = catalog['fgco2.ODIAC-FF.OCO2-MIP'].to_dask().load()
ds_odiac = xr.decode_cf(ds_odiac, decode_times=True, use_cftime=True)
ds_odiac.fgco2
```

```python
SFCO2_global_odiac = (ds_odiac.fgco2 * ds_odiac.area).sum(['lat', 'lon']) * 86400. * 365. * mwco2 * 1e-15
SFCO2_global_odiac.attrs['units'] = 'Gt CO$_2$ yr$^{-1}$'
SFCO2_global_odiac
```

```python
SFCO2_global_odiac.plot()
```

```python
SFCO2_global_odiac.plot(label='ODIAC')
SFCO2_global.sel(time=slice('1980', '2015')).plot(label='CMIP6')
plt.legend();
```
